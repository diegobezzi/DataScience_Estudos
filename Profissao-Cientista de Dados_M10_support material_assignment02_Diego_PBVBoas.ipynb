1) Com o que você viu em aula, faça uma análise das variáveis qualitativas (posse de imóvel e posse de veículo) para verificar se são boas preditoras de *renda*. Alguma delas parece discriminar *renda* de forma estatisticamente significante?

````python
# importando as bibliotecas
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
````

````python
# Carregando a base de dados e visualizando algumas informações
df = pd.read_csv("previsao_de_renda.csv")
df = df.drop('Unnamed: 0', axis=1)
df = df.drop('index', axis=1)
df.head(4)
````

````python
df_qualitativa = df[['posse_de_veiculo', 'posse_de_imovel', 'renda']]
df_qualitativa.head()
````

````python
sns.pairplot(df_qualitativa)
````

````python
# Comparando os valores
ax = sns.barplot(x="posse_de_veiculo", y="renda",
                   data=df_qualitativa, dodge=True, ci = 90)
````

````python
# Comparando os valores
sns.barplot(x="posse_de_imovel", y="renda",
                   data=df_qualitativa, dodge=True, ci = 90)
````

**Resposta**

- Ao meu ver, as variaveis __posse_de_veiculo__ e __posse_de_imovel__ não são as melhores escolhas para se fazer uma predição de renda, mesmo que estatisticamente seja possível constatar que após a análise, os proprietários de veiculo e imóvel consequentemente tem uma renda maior do que as pessoas que não possuem veiculos e imoveis.
