````python
#Importando as bibliotecas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
# Fixing random state for reproducibility
np.random.seed(2360873)
````

1) Construa um gráfico de barras para avaliar a distribuição das variáveis qualitativas no tempo. Você acha que houve alguma alteração importante a ser investigada?

````python
# Carregando a base de dados e visualizando algumas informações
df = pd.read_csv("previsao_de_renda.csv")
df = df.drop('Unnamed: 0', axis=1)
df = df.drop('index', axis=1)
df.head(4)
````

````python
df.shape
````

````python
# Filtrando os dados qualitativos da base
df_qualitativo = df[['sexo', 'posse_de_veiculo', 'posse_de_imovel', 'tipo_renda', 'educacao', 'estado_civil', 'tipo_residencia']]
````

````python
# Plotando os gráficos de contagem para todas as variáveis qualitativas
for var in df_qualitativo:
    plt.figure(figsize=(12, 6))
    plt.title(f"Distribuição de {var} ao longo do tempo")
    sns.countplot(x='data_ref', hue=var, data=df)
    plt.xticks(rotation=45)
    plt.xlabel("Data de Referência")
    plt.ylabel("Contagem")
    plt.legend(loc='upper right')
    plt.show()
````

__Resposta__

- Após análise, a única variável que teve uma variação a se destacar ao longo do tempo é a variável 'educação', que a partir de Setembro/15 houve um aumento considerável de pessoas com o superior completo e a redução de pessoas apenas com o Secundário completo. No mais, o restante das variáveis nos mostra um comportamento estável, sem qualquer tipo de destaque.

2) Para cada variável qualitativa, construa um gráfico mostrando as médias de cada nível da variável no tempo e avalie a estabilidade. Houve alguma alteração importante? Você acha que se pode utilizar essas variáveis em uma modelagem?

````python
# Selecionando as variáveis qualitativas
df_qualitativa1 = ["sexo", "posse_de_veiculo", "posse_de_imovel", "tipo_renda", "educacao", "estado_civil", "tipo_residencia"]

# Ajustando o tamanho dos gráficos e a rotação dos rótulos do eixo x
plt.figure(figsize=(60, 60))

# Plotando os gráficos de barras para cada variável qualitativa com as médias ao longo do tempo
for i, var in enumerate(df_qualitativa1, 1):
    plt.subplot(3, 3, i)
    sns.barplot(x='data_ref', y='renda', hue=var, data=df)
    plt.title(f"Médias de Renda por {var}")
    plt.xlabel("Data de Referência")
    plt.ylabel("Média de Renda")
    plt.xticks(rotation=45)
    plt.legend(loc='upper right')

plt.tight_layout()
plt.show()
````

__Resposta__

- As únicas variáveis que apresentaram algum tipo de variação relevante foram as 'tipo_renda' e 'educação'. Ao meu ver, o restante das variáveis não nos apresentam dados com profundidade relevante o suficiente para um modelo de análise.
