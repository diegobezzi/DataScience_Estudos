```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#leitura do arquivo csv com os dados do SINASC de Roraima

sinasc = pd.read_csv('SINASC_RO_2019.csv')
print(sinasc.shape)
sinasc.drop_duplicates().head()

```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
#criação de boxplot com os dados da idade da mãe

sns.boxplot(y='IDADEMAE', data=sinasc)
```

####  2) Faça um boxplot do peso do bebê

```python
#criação de boxplot com os dados do peso do bebê

sns.boxplot(y='PESO', data=sinasc)
```

####  3) Faça um histograma da idade da mãe

```python
#criação de histograma com os dados da idade da mãe

sns.displot(data = sinasc 
            ,alpha = .55
            ,x = 'IDADEMAE'
            ,bins = 20
            #,element = 'step'
            ,kde = True
           )
```

#### 4) Faça um histograma do peso do bebê

```python
#criação de histograma com os dados do peso do bebê

sns.displot(data = sinasc 
            ,alpha = .55
            ,x = 'PESO'
            ,bins = 20
            #,element = 'step'
            ,kde = True
           )
```
